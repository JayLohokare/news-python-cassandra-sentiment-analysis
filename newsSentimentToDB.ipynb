```
#Load excel sheet from server
excelLocation = "Book1.xlsx"
 
# To open Workbook
wb = xlrd.open_workbook(excelLocation)
sheet = wb.sheet_by_index(0)
 
print(sheet.nrows)
#Column 1 = API key
#Column 2 = Keyword to search
#Column 3 = Mapping coin ID
sheet.cell_value(0, 0)
```

```
def getArticleContent(url):
    article = Article(url)
    article.download()
    article.parse()
    return article.text
```

```
data = {}

for i in range(sheet.nrows):
  
  searchKey = sheet.cell_value(i, 2)
  data[searchKey] = []

  APIKey = sheet.cell_value(i, 0)
  
  coinName = sheet.cell_value(i, 1)

  baseURL = 'https://newsapi.org/v2/everything?sortBy=publishedAt&apiKey=' + APIKey + '&q=' + searchKey + '&language=en'
  print(baseURL)

  response = requests.get(baseURL)
  jsonVal = response.json()

  articlesTemp = jsonVal['articles']

  for i in articlesTemp:
    try:
      url = i['url']
      publishedAt = i['publishedAt']
      title = i['title']
      description = i['description']
      author = i['author']
      content = getArticleContent(url).strip()

      tempDict = {}
      tempDict['url'] = url
      tempDict['publishedAt'] = publishedAt
      tempDict['title'] = title
      tempDict['description'] = description
      tempDict['author'] = author
      tempDict['content'] = content
      tempDict['sentiment'] = getSentiment(content)
      tempDict['coin'] = coinName
      

      if content != "":
        data[searchKey].append(tempDict)

    except:
      continue
   
```

```
#Remove this code when going to production. 
#This just prints out all articles content
for i in data.keys():
  for j in data[i]:
    print (j['content'])
```
